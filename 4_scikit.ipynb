### 1. Getting the data ready

```python
# Import data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
# Create X (features matrix)
X = heart_disease.drop("target", axis=1)

# Create y (prediction vector)
y = heart_disease["target"]
```

### 2. Choosing appropriate model

```python
# Classification with scikit-learn
np.random.seed(0)
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100)

# See all the default parameters
clf.get_params()
```

### 3. Fit model to training data

```python
# Create train and test sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Fit classifier to the training set
clf.fit(X_train, y_train);
```

```python
# Make prediction
y_pred = clf.predict(X_test)
y_pred
```

### 4. Evaluate model

```python
# Evaluate the model using training data
clf.score(X_train, y_train)
```

```python
# Evaluate the model using test data
clf.score(X_test, y_test)
```

```python
# See metrics
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

### 5. Improve model

```python
# Try different amount of n_estimators
np.random.seed(42)
best = 0
best_n_estimators = 0
for i in range(10, 100, 10):
    print(f"Trying model with {i} estimators...")
    clf = RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
    print(
        f"Model accuracy on test set: {clf.score(X_test, y_test) * 100:.2f}%")
    if (clf.score(X_test, y_test) > best):
        best = clf.score(X_test, y_test)
        best_n_estimators = i


print(
    f"\nBest model accuracy on test set: {best * 100:.2f}% with {best_n_estimators} estimators")
```

### 6. Save and load model

```python
# Creating the model based on best number of estimators
clf = RandomForestClassifier(n_estimators=best_n_estimators).fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
import pickle

pickle.dump(clf, open('results/4_hd_classifier1.pkl', 'wb'))
```

```python
loaded_model = pickle.load(open('results/4_hd_classifier1.pkl', 'rb'))
result = loaded_model.score(X_test, y_test)
print(result)
```
