### Step by step workflow

```python
# 0. Import libraries
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(5, 5))

# 3. Plot data
ax.plot(x, y, 'o')

# 4. Customize plot
# ax.set_title('Square numbers', fontsize=24)
# ax.set_xlabel('Value', fontsize=14)
# ax.set_ylabel('Square of value', fontsize=14)
ax.set(title='Square numbers', xlabel='Value', ylabel='Square of value')

# 5. Save figure & show plot
plt.show()
fig.savefig('results/3_matplotlib.png')
```

```python
import numpy as np

x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Linear plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x), c='r')
```

```python
# Bar chart
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
# Green
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values(), color="green")
ax.set(title="Dan's Nut Butter Store", ylabel="Price ($)");
```

```python
# Horizontal bar chart
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Histogram
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# Multiples axes in one figure (Option 1)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));

fig.suptitle("Random plots")

fig.savefig("results/3_subplots.png")
```

```python
# Multiple axes in one figure (Option 2)
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000))

fig.suptitle("Random plots");
```

```python
df = pd.read_csv("heart-disease.csv")
df.head()
```

```python
df["age"].plot.hist(bins=45);
```

```python
df.plot.hist(subplots=True, figsize=(10, 30));
```

```python
over_50 = df[df['age'] > 50]
over_50.head()
```

```python
# Pyplot API method
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target');
```

```python
# Matplotlib OO method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="RdBu")

ax.set(title="Heart Disease and Cholesterol Levels",
         xlabel="Age",
            ylabel="Cholesterol");

ax.legend(*scatter.legend_elements(), title="Target")
ax.axhline(over_50["chol"].mean(),
              linestyle="--");
```

```python
over_50.head()
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)

scatter0 = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"], cmap="winter")
ax0.set(title="Heart Disease and Cholesterol Levels", ylabel="Cholesterol")
ax0.axhline(y=over_50["chol"].mean(), linestyle="--")
ax0.legend(*scatter0.legend_elements(), title="Target")

scatter1 = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"], cmap="winter")
ax1.set(title="Heart Disease and Max Heart Rate", ylabel="Max Heart Rate")
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--")
ax1.legend(*scatter1.legend_elements(), title="Target");

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
fig.savefig("results/heart_disease1.png")
```
